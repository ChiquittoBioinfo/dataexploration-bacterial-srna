```python
# Filter sequences

from Bio import SeqIO

bacterias_file = "./data/Rfam/bacteria.txt"
fasta_sequences = "./data/Rfam/raw.fa"
fasta_output = "./data/Rfam/bacteria.fa"

fasta_dict = { seq_record.id:seq_record for seq_record in SeqIO.parse(fasta_sequences, "fasta") }

lineI = 0

seqs_output = {}
with open(bacterias_file) as file:
  for line in file:
    specie_name = line.strip()

    for id, seq_record in list(fasta_dict.items()):
      if specie_name in seq_record.description:
        seqs_output[id] = seq_record
        del fasta_dict[id]
  
    lineI += 1
    if (lineI % 250) == 0:
      print(lineI, " len(fasta_dict)=", len(fasta_dict))
```

```python
with open(fasta_output, "w+") as output_file:
  for id, sequence in seqs_output.items():
    output_file.write(">" + sequence.id + "\n" + str(sequence.seq).upper() + "\n")
```
